# Streaming and Bath

- Most models can stream their output content while it is being generated. 
- By displaying output progressively, streaming significantly improves user experience, particularly for longer responses. 
- Calling `stream()` returns an iterator thst yields output chunks as they are produced.
- You can use a loop to process each chunk in real-time:  

```python
model.invoke("Write me a 200 words paragraph on Artificial Intelligence") # we have to wait till the llm complete its output
```

```python
# streaming = 
for chunk in model.stream("Write me a 200 words paragraph on Artificial Intelligence"):
    print(chunk.text, end="|", flush=True) 
```

```python
# streaming = 
for chunk in model.stream("Why do parrots have colorful feather"):
    print(chunk.text, end="|", flush=True) 
```

```python
# It takes a very long time. 
model.invoke("Why do parrots have colorful feather")
```

## Batch
- Batching is a collection of independent requests to a model can significantly improve performance and reduce costs, as the processing can be done in **parallel**

```python
# send all the questions as an input to the llm model in a parallel way and ouput parallel
responses = model.batch([
    "Why do parrots have colorful feathers?",
    "How do airplanes fly?",
    "What is quantum computing?"
])

for response in responses:
    print(response)
```

```python
# send all the questions as an input to the llm model in a parallel way and ouput parallel
responses = model.batch([
    "Why do parrots have colorful feathers?",
    "How do airplanes fly?",
    "What is quantum computing?" ],
    config={
        "max_concurrency": 5  # Limit 5 bath parallel
    }
)

for response in responses:
    print(response)
```
