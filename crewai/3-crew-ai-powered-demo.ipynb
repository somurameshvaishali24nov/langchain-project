    name: str = "Web Search"
    description: str = "Search the web for recent information."

    def _run(self, query: str):

            "api_key": TAVILY_API_KEY,

````python
# Step 2: Import the required libraries

import os
import getpass
import requests

from crewai import Agent, Task, Crew, Process, LLM
from crewai.tools import BaseTool
````

## Step 3: Securely enter API keys

````python
# The entered values are hidden and are not saved inside the notebook.

os.environ["ALOK_OPEN_API_KEY"] = os.getenv("ALOK_OPEN_API_KEY")
os.environ["TAVILY_API_KEY"] = os.getenv("TAVILY_API_KEY")

if not os.environ["ALOK_OPEN_API_KEY"].strip():
    raise ValueError("OpenAI API key cannot be empty.")

print("os.getenv('TAVILY_API_KEY'): ", os.getenv("TAVILY_API_KEY"))
if not os.environ["TAVILY_API_KEY"].strip():
    raise ValueError("Tavily API key cannot be empty.")

print("API keys configured successfully.")
````

## Step 4: Create a custom Tavily web-search tool

````python


class TavilySearchTool(BaseTool):
    name: str = "Tavily Web Search"
    description: str = (
        "Search the web for recent and reliable information. "
        "Provide a focused search query as input."
    )

    def _run(self, query: str) -> str:
        if not query or not query.strip():
            return "Search query cannot be empty."

        url = "https://api.tavily.com/search"
        payload = {
            "api_key": os.environ["TAVILY_API_KEY"],
            "query": query,
            "search_depth": "advanced",
            "max_results": 5,
            "include_answer": False,
        }

        try:
            response = requests.post(url, json=payload, timeout=30)
            response.raise_for_status()
            data = response.json()
        except requests.RequestException as exc:
            return f"Web search failed: {exc}"
        except ValueError:
            return "Web search failed because Tavily returned an invalid response."

        results = data.get("results", [])
        if not results:
            return "No relevant web-search results were found."

        formatted_results = []
        for index, result in enumerate(results, start=1):
            title = result.get("title", "Untitled result")
            url = result.get("url", "URL unavailable")
            content = result.get("content", "No summary available")
            formatted_results.append(
                f"{index}. {title}\n"
                f"URL: {url}\n"
                f"Summary: {content}"
            )

        return "\n\n".join(formatted_results)


search_tool = TavilySearchTool()
print("Tavily search tool created.")
````

````python
# Step 5: Configure the standard OpenAI model
# Change the model name below when you want to use another model
# available in your OpenAI project.

llm = LLM(
    model="openai/gpt-4.1-nano",
    api_key=os.environ["OPENAI_API_KEY"],
    temperature=0.3,
)

print("OpenAI LLM configured.")
````

### Step 6: Define the CrewAI agents

````python
researcher = Agent(
    role="AI Researcher",
    goal=(
        "Find and explain the three most important recent advancements "
        "in artificial intelligence for the FMCG industry."
    ),
    backstory=(
        "You are an AI industry researcher who specializes in FMCG, "
        "retail, supply chains, demand forecasting, personalization, "
        "marketing, and intelligent automation. You use web search to "
        "collect current and verifiable information."
    ),
    llm=llm,
    tools=[search_tool],
    verbose=True,
    allow_delegation=False,
)

writer = Agent(
    role="Technical Writer",
    goal=(
        "Transform the research findings into a clear and concise "
        "executive report for business leaders."
    ),
    backstory=(
        "You are an experienced technical writer who converts complex "
        "AI research into business-friendly executive summaries."
    ),
    llm=llm,
    verbose=True,
    allow_delegation=False,
)

print("Agents created successfully.")
````

### Step 7: Define the agents' tasks

````python
task_research = Task(
    description=(
        "Search the web and identify the top three recent advancements "
        "in AI for the FMCG industry. For each advancement, explain the "
        "technology, its business use, a practical example, and the source URL. "
        "Prioritize recent and credible information."
    ),
    expected_output=(
        "Detailed research notes covering exactly three recent AI advancements "
        "in FMCG, including business impact, examples, and source URLs."
    ),
    agent=researcher,
)

task_write = Task(
    description="""
Using the research notes, write an executive summary.

Requirements:
- Maximum 200 words
- Use exactly three bullet points
- Cover only the three selected advancements
- Explain the business value of each advancement
- Include source URLs supplied by the researcher
- Do not invent facts or sources
""",
    expected_output=(
        "A concise executive summary containing exactly three bullet points "
        "and no more than 200 words."
    ),
    agent=writer,
    context=[task_research],
)

print("Tasks created successfully.")
````

### Step 8: Assemble and run the crew

````python
crew = Crew(
    agents=[researcher, writer],
    tasks=[task_research, task_write],
    process=Process.sequential,
    verbose=True,
)

result = await crew.kickoff_async()

print("\n" + "=" * 70)
print("FINAL OUTPUT")
print("=" * 70)
print(result.raw)
````
